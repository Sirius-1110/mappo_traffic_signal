# 类 DiscreteHead 的作用
DiscreteHead 类是一个用于输出离散动作 logit 的神经网络头部组件。具体来说，它接受一个形状为 (B, N) 的输入张量，并返回一个包含输出 logit 的字典。logit 是每个可能离散动作的概率分布的对数形式。

主要功能
- 输入处理：
输入是一个形状为 (B, N) 的张量，其中 B 是批量大小，N 是隐藏层的大小。

- 多层感知机 (MLP) 处理：
使用多层感知机（MLP）对输入进行处理。MLP 的层数、激活函数和归一化类型可以根据初始化参数进行配置。
如果 noise 参数设置为 True，则使用 NoiseLinearLayer 作为线性层，否则使用标准的 nn.Linear 层。
- 输出生成：
最后一层是一个全连接层，将 MLP 的输出映射到 output_size 维的 logit 张量。
- 返回
一个字典，其中包含键 'logit'，其值为生成的 logit 张量。

```python
import torch
import torch.nn as nn
from ding.model.common.head import DiscreteHead
from ding.torch_utils import MLP, fc_block, NoiseLinearLayer, noise_block

def test_discrete_head():
    # 测试参数
    batch_size = 4
    hidden_size = 64
    output_size = 10
    layer_num = 2
    activation = nn.ReLU()
    norm_type = 'LN'  # Layer Normalization
    noise = False

    # 初始化 DiscreteHead
    discrete_head = DiscreteHead(
        hidden_size=hidden_size,
        output_size=output_size,
        layer_num=layer_num,
        activation=activation,
        norm_type=norm_type,
        noise=noise
    )

    # 生成随机输入
    inputs = torch.randn(batch_size, hidden_size)

    # 前向传播
    outputs = discrete_head(inputs)

    # 验证输出
    assert isinstance(outputs, dict), "输出应为字典"
    assert 'logit' in outputs, "输出字典应包含键 'logit'"
    assert isinstance(outputs['logit'], torch.Tensor), "'logit' 应为张量"
    assert outputs['logit'].shape == (batch_size, output_size), f"'logit' 形状应为 ({batch_size}, {output_size})"

    print("所有测试通过！")

# 运行测试
test_discrete_head()
```
